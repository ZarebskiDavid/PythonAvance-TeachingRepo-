### Restons d'abord sur Quandl 

Fort de votre succès, le marketeux devient plus greedy et vous demande un moyen de prédire l'évolution du cours sur la base des variations déjà enregistrées. Nous entrons là dans un monde nouveau: celui des <b>régressions</b> et du <b>machine learning</b>. 

Nous allons utiliser un autre data set pour cet exercice: 'WIKI/GOOGL'

Il existe une multitude de manières (i.e. d'algorithmes) pour prédire les variations d'une variable. Nous allons commencer simplement avec un bête <a href="https://fr.wikipedia.org/wiki/R%C3%A9gression_lin%C3%A9aire">model linéaire</a>. La librairie de référence pour ce type de taches en python s'appelle la <a href="http://scikit-learn.org/stable/">Scikit-learn</a>. 

![Model linéaire](http://scikit-learn.org/stable/_images/sphx_glr_plot_ols_001.png)

L'idée générale à retenir est la suivante. En machine learning, le but consiste à trouver une fonction capable de modéliser la relation entre deux variables X Y. (e.g. on peut s'attendre à ce que la relation entre le poids et la taille dans une population soit adéquatement rempli par une fonction linéaire). l'idée est d'entrainer l'algorithme sur une subset de donnée X_train Y_train et de voir si les paramètres retenus s'appliquent adéquatement au reste.

<ul>
<li> nous définissons X_train Y_train 
<li> nous fittons le modèle
<li> le modèle doit arriver à estimer Y_test sur la base de X_test
</ul>

Au commencement était le df

```python
df = quandl.get('WIKI/GOOGL')

df = df[['Adj. Open', 'Adj. High', 'Adj. Low', 'Adj. Close', 'Adj. Volume']]
df['HL_PCT'] = (df['Adj. High'] - df['Adj. Close']) / df['Adj. Close'] * 100.0
df['PCT_change'] = (df['Adj. Close'] - df['Adj. Open']) / df['Adj. Open'] * 100.0
df
```

```python
df.to_json("quandl.json")
```

```python
import math, datetime, time

forecast_col = 'Adj. Close'
df.fillna(-9999, inplace=True)

forecast_out = int(math.ceil(0.01*len(df)))
print(forecast_out)

df['label'] = df[forecast_col].shift(-forecast_out)
df
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

X = np.array(df.drop(['label'],1))
X = preprocessing.scale(X)
X_lately = X[-forecast_out:]
X = X[:-forecast_out]

print(X)
print(X.shape)

df.dropna(inplace=True)
Y = np.array(df['label'])

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

print(Y)
print(Y.shape)
```

```python
from sklearn.linear_model import LinearRegression
import pickle

clf = LinearRegression(n_jobs=-1)
clf.fit(X_train, Y_train) #training
with open('linearregression.pickle', 'wb') as f: 
    pickle.dump(clf, f)
# training: = > comment when you have your model

pickle_in = open ('linearregression.pickle', 'rb')
clf = pickle.load(pickle_in)

accuracy = clf.score(X_test,Y_test)
forecast_set = clf.predict(X_lately)

print(accuracy)
```

### Essayons autre chose: taxis!

La finance, c'est bien gentil mais mettons que nous voulions vraiment aider les gens. Les csv trainTaxi et testTaxi contiennent les données de près d'un million et demi de trajets de taxi new-yorkais. Nous voulons entrainer notre algorithme à deviner le temps de trajet des courses sur la base des données disponibles. Regardons d'abord ce dernières:


```python
train = pd.read_csv('trainTaxi.csv')
test = pd.read_csv('testTaxi.csv')
train.head(20)
```

#### D'abord, on clean



```python
# new york city area
min_lon = -74.844,
min_lat = 40.026
max_lon = -72.221
max_lat = 41.372

initial_len = train.shape[0]

# ruling points outside area as outliers
train = train[train['pickup_longitude'].between(min_lon, max_lon)]
train = train[train['pickup_latitude'].between(min_lat, max_lat)]
train = train[train['dropoff_longitude'].between(min_lon, max_lon)]
train = train[train['dropoff_latitude'].between(min_lat, max_lat)]

cleaned_len = train.shape[0]
initial_len - cleaned_len
```

```python
train = train.drop('id', axis=1)
train.isnull().sum()
```

#### Encoder le temps
Les dates et heures apparaissent pour le moment sous la forme de chaine de caractères. Il serait bon de les distinguer dans des champs de données à part

```python
# to datetime
train['pickup_datetime'] = pd.to_datetime(train['pickup_datetime'])
train['dropoff_datetime'] = pd.to_datetime(train['dropoff_datetime'])
```

```python
# get day and hour information
train['pickup_hour'] = train['pickup_datetime'].apply(lambda x: x.hour)
train['pickup_day'] = train['pickup_datetime'].apply(lambda x: x.weekday())
```

```python
# convert timestamps to float
train['pickup_datetime'] = train['pickup_datetime'].apply(lambda x: x.timestamp())
train['dropoff_datetime'] = train['dropoff_datetime'].apply(lambda x: x.timestamp())
```

```python
# get dummy variables for categorial data
train['store_and_fwd_flag'] = pd.get_dummies(train['store_and_fwd_flag'])
```

Nous allons nous servir d'un algorithme de type random tree forest (les détails ne nous intéresse ici pas au premier chef mais nous pouvez vous faire une idée de la logique sous jacente de cette algorithme à l'aide de <a href ="https://youtu.be/D_2LkhMJcfY"> cette vidéo </a>  et, bien sur, <a href ="https://en.wikipedia.org/wiki/Random_forest"> la page wiki correspondante </a>). Par chance, cet algorithme est dejà implémenté dans la librairie <a href="http://scikit-learn.org/stable/modules/ensemble.html#forests-of-randomized-trees"> sklearn </a>

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import normalize
from sklearn.model_selection import train_test_split
```

```python
X = train.drop('trip_duration', axis=1)
y = train['trip_duration']
X
```

```python
X = normalize(X)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```
